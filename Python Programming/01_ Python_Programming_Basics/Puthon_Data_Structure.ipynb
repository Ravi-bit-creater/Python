### Four types of Data Types

2. Sets

## 1.List
- Denoted by square bracket [].
- List is 1D
- It can handle multiple data types and it is a hetrogenous data structure
- Allows Duplicate Values
- Indexing or Slicing is possible
- Mutable [Values once assigned to a variable it can be changed]

```python
empty_cart = [] # Empty List
item_to_buy = ["Apple", "Orange", "Mango", "Banana","Apple"]
len(item_to_buy) # Length of the list
item_to_buy # Allows Duplicate elements
```

```python
item_to_buy[1:5] # Indexing using positive number such as left to Right
```

```python
item_to_buy[-3:-1] # indexing using the negavtive numbers such as Right to Left 
```

```python
item_to_buy[1] = "Watermelon"
item_to_buy #Changing the element in the list
```
