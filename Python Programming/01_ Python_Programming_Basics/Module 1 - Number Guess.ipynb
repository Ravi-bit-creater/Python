## Level 2 : Print if the user guess number is wright or not with the 7 trails

#### Level 2.1 : For Loop - It is used when the end number is defined 

```python
secert_number = 63  # Declaring a variable

for i in range(1,8): # It will run until 7 times
    user_input = int(input("Enter your guess:"))     # Receiving the input from the user as an interger using  int keyword
    
    if secert_number == user_input:
        print("Congradulation! You're guess is correct")
        break   # is used to break the statment when the above condition gets true
    else:
        print("Sorry. You're guess is wrong")
```

 #### Level 2.2 : While Loop - It is used when the end number is not defined and when to run the loop infinite times

```python
secert_number = 63  # Declaring a variable

while True: # It will run infinite times until the condition gets true
    user_input = int(input("Enter your guess:"))     # Receiving the input from the user as an interger using  int keyword
    
    if secert_number == user_input:
        print("Congradulation! You're guess is correct")
        break
    else:
        print("Sorry. You're guess is wrong")
```

## Level 6 : How to handle invalid characters or values

```python
import random

print("Welcome to Guess the Number Challange !!! ")
print("Enter the number from 1-100")

secert_number = random.randint(1,100)

for i in range(1,8): # creating a for loop to run 7 times 

    try:
        user_input = int(input("Enter the number:"))
        
        # Nested if is used to here for printing the number is higher or lower or correct
        if user_input < 1 or user_input > 100:
            print("Invalid Number!, Please enter the number between 1-100")
        elif user_input > secert_number:
            print("The number you have guess is higher")
        elif user_input < secert_number:
            print("The number you have guess is lower")
        else:
            print("Congradulations! The number you have guess is correct")
            break
    except ValueError:
        print("Invalid Number, Not Recoganized. Please enter the number between 1-100")
        
if secert_number == user_input:
    print(f"You have guessed the number {secert_number} in the {i} attempts") # print formatting is used here to print the secert number & attempts
else:
    print("Sorry your attempts have exceeds.. Try Again !!")
```

## Level 7 : Take this program to run as a Function when it is called.
1. Define a function
2. Call the function

```python
def number_guessing_game():  # Define the function
    import random
    
    print("Welcome to Guess the Number Challange !!! ")
    print("Enter the number from 1-100")
    
    secert_number = random.randint(1,100)
    
    for i in range(1,8): # creating a for loop to run 7 times 
    
        try:
            user_input = int(input("Enter the number:"))
            
            # Nested if is used to here for printing the number is higher or lower or correct
            if user_input < 1 or user_input > 100:
                print("Invalid Number!, Please enter the number between 1-100")
            elif user_input > secert_number:
                print("The number you have guess is higher")
            elif user_input < secert_number:
                print("The number you have guess is lower")
            else:
                print("Congradulations! The number you have guess is correct")
                break
                
        except ValueError:   # It will print this when the value error occurs
            print("Invalid Number, Not Recoganized. Please enter the number between 1-100")
            
    if secert_number == user_input:
        print(f"You have guessed the number {secert_number} in the {i} attempts") # print formatting is used here to print the secert number & attempts
    else:
        print("Sorry your attempts have exceeds.. Try Again !!")

number_guessing_game() # Call the function
```
