### 1. Check even or odd
Input a number → print “Even” or “Odd”.

```python
user_input = int(input("Enter the number"))

if user_input % 2 == 0:
    print(f"The number {user_input} is Even")
else:
    print(f"The number {user_input} is Odd")
```

### 2. Find the biggest among 3 numbers
Use if-else.

```python
num_1 = int(input("Enter the number 1"))
num_2 = int(input("Enter the number 2"))
num_3 = int(input("Enter the number 3"))

if num_1 == num_2 == num_3:
    print("Kindly enter the different numbers  should not be the same")
elif num_1 > num_2 and num_1 > num_3:
    print(f"The 1st number {num_1} is bigger")
elif num_2 > num_1 and num_2 > num_3:
    print(f"The 2nd number {num_2} is bigger")
else:
    print(f"The 3rd number {num_3} is bigger")
```

## 24. Print the sum of digits of a 2-digit number
Example: 47 → 4 + 7 = 11.

```python
num = int(input("Enter the number"))
digit_sum = 0 
num_dup = num

while num > 0:
    digit = num % 10  # Extract the last digit
    digit_sum += digit
    num //= 10  # Removes the last digit
print("The sum of ",num_dup," is: ", digit_sum)
```

## 25.Print all numbers from 1 to N, but stop when the number reaches 10
Use a while loop.
Example: If N = 25 → it should print only 1 to 10.

```python
n = int(input("Enter the number"))
i = 1
while i <= n:
    print(i)
    if i == 10:
        break
    i += 1
```

## 27.Find the youngest age among three people

```python
age_1 = int(input("Enter the age 1"))
age_2 = int(input("Enter the age 2"))
age_3 = int(input("Enter the age 3"))

if age_1 < age_2 and age_1 < age_3:
    print("Age 1 is youngest")
elif age_2 < age_1 and age_2 < age_3:
    print("Age 2 is youngest")
elif age_3 < age_1 and age_3 < age_2:
    print("Age 3 is youngest")
else:
    print("There is a tie in the youngest age")
```
