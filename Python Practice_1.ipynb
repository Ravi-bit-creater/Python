## 64. Question 4: Password Validator

Write a program to repeatedly ask the user to enter a password until it meets the rules:

At least 8 characters

Contains a number

Contains both uppercase and lowercase letters
(Use while loop + if conditions)
When valid, print "Password accepted!".

```python
while True:
    password = input("Enter the password")
    
    if len(password) < 8:
        print("The password must contains more than 8 digit")
        continue

    if not any(char.isdigit() for char in password):
        print("The password must contains a digit")
        continue

    if not any(char.islower() for char in password):
        print("The password must contains a lower case character")
        continue

    if not any(char.isupper() for char in password):
        print("The password must contains a upper case character")
        continue

    print("Password Accpected")
    break
    
```

## 70. Character Counter with Retry

Write a program that repeatedly asks the user to enter a sentence.

Rules:

Use a while loop to continue asking until the sentence length is at least 15 characters.

Once a valid sentence is entered:

Use a for loop to count:

Number of uppercase letters

Number of lowercase letters

Number of digits

Number of spaces

Use if conditions inside the loop to classify each character.

```python
while True:
    sentence = input("Enter a sentence (minimum 15 characters): ")
    
    if len(sentence) >= 15:
        break
    else:
        print("Sentence too short! Please try again.\n")

uppercase = 0
lowercase = 0
digits = 0
spaces = 0

for ch in sentence:
    if ch.isupper():
        uppercase += 1
    elif ch.islower():
        lowercase += 1
    elif ch.isdigit():
        digits += 1
    elif ch.isspace():
        spaces += 1

print("\nValid sentence accepted.")
print("Uppercase:", uppercase)
print("Lowercase:", lowercase)
print("Digits:", digits)
print("Spaces:", spaces)
```

## 122. Smart Warehouse Load Balancer
Concept

Optimization + threshold logic

Question

Store weight loads for 5 warehouses.

Max capacity = 1000 units.

If load > capacity → "Overloaded"

Redistribute excess equally to underloaded warehouses.

```python
loads = []
capacity = 1000

for i in range(5):
    loads.append(int(input(f"Warehouse {i+1} load: ")))

excess = 0
for i in range(5):
    if loads[i] > capacity:
        excess += loads[i] - capacity
        loads[i] = capacity

underloaded = [i for i in range(5) if loads[i] < capacity]
share = excess // len(underloaded) if underloaded else 0

for i in underloaded:
    loads[i] += share

print("Final Loads:", loads)
```

If all rooms cleared → Victory 🎉

If player quits → Exit game 🚪

```python
import random

# Player stats
health = 100
energy = 50
rooms = 5

print("🏰 Welcome to the Dungeon Survival Challenge!")
print("You must survive all rooms to escape.\n")

room_number = 1

# Main game loop
while health > 0 and room_number <= rooms:
    print(f"\n--- Room {room_number} ---")
    event = random.choice(["enemy", "trap", "treasure"])

    # Enemy encounter
    if event == "enemy":
        print("🧟 An enemy appears!")
        if energy >= 10:
            print("You fight the enemy.")
            energy -= 10
            health -= random.randint(5, 15)
        else:
            print("Too tired to fight! Enemy attacks you.")
            health -= random.randint(15, 25)

    # Trap encounter
    elif event == "trap":
        print("⚠️ You triggered a trap!")
        health -= random.randint(10, 20)

    # Treasure encounter
    else:
        print("💰 You found a treasure!")
        reward = random.choice(["health", "energy"])
        if reward == "health":
            gain = random.randint(10, 20)
            health += gain
            print(f"Health increased by {gain}")
        else:
            gain = random.randint(10, 20)
            energy += gain
            print(f"Energy increased by {gain}")

    # Player status
    print(f"Health: {health}")
    print(f"Energy: {energy}")

    # Player choice
    choice = input("Do you want to continue? (yes/no): ")
    if choice.lower() == "no":
        print("🚪 You exited the dungeon.")
        break

    room_number += 1

# Game result
if health <= 0:
    print("\n💀 Game Over! You did not survive the dungeon.")
elif room_number > rooms:
    print("\n🎉 Congratulations! You escaped the dungeon successfully!")
```
